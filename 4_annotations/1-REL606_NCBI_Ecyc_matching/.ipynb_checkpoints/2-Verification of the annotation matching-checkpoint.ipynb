# Verification

```python
import numpy as np
import pandas as pd
```

```python
ann_file = "Annotation_NCBI_Ecocyc.csv"
ann_match = pd.read_csv(ann_file, header = 0, index_col = 0)
ann_match.head()
```

```python
ann_match.columns
```

```python
# calculate the distance
distance = np.abs(np.asarray(ann_match["START-BASE"])-np.asarray(ann_match["start"]))+np.abs(np.asarray(ann_match["END-BASE"])-np.asarray(ann_match["end"]))
ann_match["distance"] = distance
ann_match.sort_values(by = ["distance"], ascending=False)
```

```python
# calculate if there are overlap.
overlap_list = []
eco_len_list = []
gen_len_list = []
def overlap_calc(min1, max1, min2, max2):
    "Description: this is the magic."
    return max(0, min(max1, max2) - max(min1, min2))

i = 0
while i<len(ann_match):
    eco_start = ann_match.iloc[i]["START-BASE"]
    eco_end = ann_match.iloc[i]["END-BASE"]
    gen_start = ann_match.iloc[i]["start"]
    gen_end = ann_match.iloc[i]["end"]
    
    eco_len = abs(eco_end-eco_start)
    gen_len = abs(gen_end-gen_start)
    
    min1 = min([eco_start, eco_end])
    max1 = max([eco_start, eco_end])
    
    min2 = min([gen_start, gen_end])
    max2 = max([gen_start, gen_end])
    
    overlap = overlap_calc(min1, max1, min2, max2)
    overlap_list.append(overlap)
    eco_len_list.append(eco_len)
    gen_len_list.append(gen_len)
        
    i = i+1
    
ann_match["overlap"] = overlap_list
ann_match["ecocyc_gene_len"] = eco_len_list
ann_match["NCBI_gene_len"] = gen_len_list
```

```python
ann_match.sort_values(by = ["distance"], ascending=False).to_csv("Annotation_NCBI_Ecocyc_with_dist.csv")
```
