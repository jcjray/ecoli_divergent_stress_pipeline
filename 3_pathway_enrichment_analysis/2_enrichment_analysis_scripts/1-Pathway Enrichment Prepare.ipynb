# Pathway Enrichment Prepare

This file mainly create a .csv file listing all pathways in rel606, and corresponding ECB_RS in that pathway.

# 1. Load

```python
import pandas as pd
import numpy as np
import Ecocyc_Parse as ep
import cobra
import matplotlib.pyplot as plt
pd.options.display.max_rows = 99
pd.options.display.max_columns = 999
```

```python
pathway_file = "HelpFiles/pathways.col"
clean_ann_file = "HelpFiles/Refined_Final_Annotation_NCBI_Ecocyc_with_dist.csv"
```

```python
pathway = ep.parse_col(pathway_file)
pathway.head(3)
```

```python
clean_ann = pd.read_csv(clean_ann_file, header = 0, index_col= 0)
clean_ann.head(3)
```

# 2. Create Pathway Dataframe.

### 2.1 Select pathway dataframe for replacing

```python
pathway.columns
```

```python
# select columns that only have "GENE-ID" in it.
idx_list = [0,1]
i = 0
for col in pathway.columns:
    if "GENE-ID" in col:
        idx_list.append(i)
    i = i+1
```

```python
# partial pathway with gene IDs
pathway_part = pathway.iloc[:, idx_list].copy(deep=True)
pathway_part.head(3)
```

### 2.2 Create gene ID -- ECB_RS## dict

```python
clean_ann.head(3)
```

```python
# replace the ecycID as ECB_RS#
clean_ann_part = clean_ann[["UNIQUE-ID", "locus_tag"]]
clean_ann_part.head(3)
```

```python
# gene id -- ECB_RS#
genid_ecbrs_dict = pd.Series(clean_ann_part["locus_tag"].values, index=clean_ann_part["UNIQUE-ID"]).to_dict()
genid_ecbrs_dict
```

### 2.3 replace the selected pathway dataframe with the ECB_RS#

```python
pathway_part.replace(genid_ecbrs_dict, inplace=True)
pathway_part.head(3)
```

```python
pathway_part.shape
```

```python
# double check if there exist unreplace ecyc id
sum_unreplace = 0
for col in pathway_part.columns:
    unreplace = pathway_part[col].str.contains("GCQD-").sum()
    sum_unreplace = sum_unreplace+unreplace
    
    if unreplace != 0:
        # If there are unreplaced genes, print the record.
        print(pathway_part[pathway_part[col].str.contains("GCQD-") == True])
print(sum_unreplace)
```

Great! So all genes are properly replaced.

### 2.4 Create merged columns for genes

```python
# create a merged column
pathway_part['gene_all'] = pathway_part[pathway_part.columns[2:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
pathway_part.head(3)
```

```python
pathway_info = pathway_part[["UNIQUE-ID", "NAME", "gene_all"]].copy(deep=True)
pathway_info.head()
```

```python
pathway_info.shape
```

```python
pathway_info.to_csv("HelpFiles/Ecyc_NCBI_pathways.csv")
```
